```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from Preprocessor import Preprocessor

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score

import xgboost as xgb
from xgboost import XGBClassifier
from xgboost import plot_importance

import datetime
today = str(datetime.datetime.now().date()).replace("-","")
print("오늘 날짜 : " + today)
```

```python
preprocessor = Preprocessor()
```

```python
train_df = preprocessor.preprocessing('../data/train.csv')
test_df  = preprocessor.preprocessing('../data/test.csv')
```

```python
X = train_df.drop('credit',axis=1)
y = train_df['credit']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1234)
```

```python
X_train = X_train.reset_index().drop(['index'],axis=1)
y_train = y_train.reset_index().drop(['index'],axis=1)
```

```python
X_train.head()
```

```python
xgb_wrapper = XGBClassifier(n_estimators=5000, learning_rate=0.01, max_depth=12)
evals = [(X_test, y_test)]
xgb_wrapper.fit(X_train, y_train, early_stopping_rounds=50, eval_metric="mlogloss", 
                eval_set=evals, verbose=True)
```

# 후진소거법으로 최고 점수 받은 변수

```python
important_columns = ['gender', 'car', 'reality', 'income_total', 'edu_type', 'family_type', 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'work_phone', 'phone', 'email', 'family_size', 
 'begin_month', 'house_type_House / apartment', 'income_type_Commercial associate', 'income_type_Working','house_type_With parents', 'income_type_State servant']
```

```python
X_train_a = X_train[important_columns]
X_test_a = X_test[important_columns]

xgb_wrapper = XGBClassifier(n_estimators=5000, learning_rate=0.01, max_depth=12)
evals = [(X_test_a, y_test)]
xgb_wrapper.fit(X_train_a, y_train, early_stopping_rounds=50, eval_metric="mlogloss", 
                eval_set=evals, verbose=True)
```

```python
test_df = test_df[important_columns]
preds = xgb_wrapper.predict_proba(test_df)
preds = pd.DataFrame(preds)

index = pd.DataFrame(range(26457,36457),columns=["index"])
preds= pd.concat([index, preds], axis=1)

preds.to_csv('../submission/'+today+'_xgb_sub.csv', index =False)
```
