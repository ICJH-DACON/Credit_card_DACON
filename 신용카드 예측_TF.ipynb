```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from Preprocessor import Preprocessor

from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, BatchNormalization
from tensorflow.keras.utils import to_categorical

import datetime
today = str(datetime.datetime.now().date()).replace("-","")
print("오늘 날짜 : " + today)
```

```python
preprocessor = Preprocessor()
```

```python
train_df = preprocessor.preprocessing('../data/train.csv')
test_df  = preprocessor.preprocessing('../data/test.csv')
```

```python
X = train_df.drop('credit',axis=1)
y = train_df['credit']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1234)
```

```python
X_tf_train = np.array(X_train)
X_tf_test = np.array(X_test)
X_tf_train = X_tf_train.reshape(-1,1,24)
X_tf_test = X_tf_test.reshape(-1,1,24)
y_tf_train = to_categorical(y_train)
y_tf_test = to_categorical(y_test)
```

```python
X_tf_train.shape, y_tf_train.shape, X_tf_test.shape, y_tf_test.shape
```

```python
model = Sequential()
model.add(BatchNormalization(input_shape=(1,24)))
# model.add(Dense(128, activation='relu'))
model.add(LSTM(64,input_shape=(1,24))) # RNN 셀의 hidden_size는 32
# model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
history = model.fit(X_tf_train, y_tf_train, epochs=200, batch_size=32)
```

```python
test_df = np.array(test_df)
test_df = test_df.reshape(-1,1,24)
```

```python
preds = model.predict(test_df)
preds = pd.DataFrame(preds)

index = pd.DataFrame(range(26457,36457),columns=["index"])
preds= pd.concat([index, preds], axis=1)

preds.to_csv('../submission/'+today+'_tf_sub.csv', index =False)
```
